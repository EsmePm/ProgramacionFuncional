```python
import csv
```

### 1. Crear una lista de diccionarios con todos los registros y todas columnas del csv

```python
def read_all(name_file):
    with open(name_file) as csvfile:
        list_worlds = []
        worlds = csv.reader(csvfile, delimiter=',')
        h = next(worlds)
        for world in worlds:
            iterador = zip(h,world)
            list_worlds.append({clave: valor for clave, valor in iterador})
        return list_worlds

l = read_all('./world_population.csv')
print(l)
```

### 2. Crear una lista de diccionarios con todos los registros y todas columnas del csv, excepto: 2022 Population, 2020 Population, 2015 Population, 2000 Population, 1990 Population, 1980 Population, 1970 Population, todos esto datos excluidos deben ser agregados a cada diccionario en una lista con la clave: Population

```python
def read_population(name_file):
    with open(name_file) as csvfile:
        list_worlds = []
        worlds = csv.reader(csvfile, delimiter=',')
        h = next(worlds)
        del h[5:13]
        for world in worlds:
            list_aux = []
            iterador = zip(h,world)
            l = {clave: valor for clave, valor in iterador}
            for a in range(5,13):
                list_aux.append(world[a])

            l['Population'] = list_aux
            list_worlds.append(l)
        return list_worlds

b = read_population('./world_population.csv')
print(b)
```

### 3. Crear una lista de diccionarios con todos los registros y con las columnas: Country/Territory y 2022 Population.

```python
def read_country(name_file):
    with open(name_file) as csvfile:
        list_worlds = []
        worlds = csv.reader(csvfile, delimiter=',')
        h = next(worlds)
        for world in worlds:
            iterador = zip(h,world)
            list_worlds.append({clave: valor for clave, valor in iterador if clave == 'Country/Territory' or clave=='2022 Population'})
        return list_worlds
        
l = read_country('./world_population.csv')
print(l)
```

### 4. Crear una lista de tuplas , donde las tuplas están compuestas por las columnas: CCA3 y World Population Percentage de solo aquellos países que tienen un World Population Percentage mayor o igual a 1.0

```python
def read_CCA3(name_file):
    with open(name_file) as csvfile:
        worlds = csv.reader(csvfile, delimiter=',')
        list_worlds = []
        h = next(worlds)
        for i in worlds :
            if float(i[16]) >= 1.0:
                list_worlds.append(tuple([i[1], float(i[16])]))
        return list_worlds
        
   
l = read_CCA3('./world_population.csv')
print(l)
```

### 5. Crear una lista de tuplas , donde las tuplas están compuestas por las columnas: CCA3 y Avg Population, este último valor se debe obtener del promedio de: 2022 Population, 2020 Population, 2015 Population, 2000 Population, 1990 Population, 1980 Population, 1970 Population.

```python
from statistics import mean

def read_avg(name_file):
    with open(name_file) as csvfile:
        list_worlds = []
        worlds = csv.reader(csvfile, delimiter=',')
        h = next(worlds)
        for world in worlds:
            list_aux = []
            for i in range(5,13):
                list_aux.append(int(world[i]))
            aux = mean(list_aux)
            list_worlds.append(tuple([world[1], aux]))
        return list_worlds

l = read_avg('./world_population.csv')
print(l)
```

### 6. Crear un conjunto de todos los países que sus elementos están formados una tupla: Country/Territory y Continent

```python
def read_territory(name_file):
    with open(name_file) as csvfile:
        worlds = csv.reader(csvfile, delimiter=',')
        h = next(worlds)
        t = set() #set ayuda a crear un conjunto vacio
        for world in worlds:
            iterador = zip(h,world)
            t.add(tuple([valor for clave, valor in iterador if clave == 'Country/Territory' or clave =='Continent']))
        return t

t = read_territory('./world_population.csv')
print(t)
```

### 14. Mediante operación de conjuntos unir los conjuntos del ejercicio 11, 12, 13.

```python
def read_union(name_file):
    with open(name_file) as csvfile:
        worlds = csv.reader(csvfile, delimiter=',')
        h = next(worlds)
        t = set()
        a = set()
        e = set()
        for i in worlds:
            if i[4]=='Africa':
                a.add(tuple([i[2],i[4]]))
            if i[4]=='Europe':
                e.add(tuple([i[2],i[4]]))
            if i[4]=='Asia':
                t.add(tuple([i[2],i[4]]))
        a.update(e)
        a.update(t)
        return a
        
l = read_union('./world_population.csv')
print(l)
```

### 18. Mediante operación de conjunto obtener los países que se encuentran tanto en el conjunto del ejercicio 16 como en el conjunto del ejercicio 17.

```python
def read_union(name_file):
    with open(name_file) as csvfile:
        worlds = csv.reader(csvfile, delimiter=',')
        h = next(worlds)
        t = set()
        a = set()
        s = set()
        for i in worlds:
            if float(i[16]) > 0.2:
                t.add(i[2])
            if float(i[16]) < 0.4:
                a.add(i[2])
        s = t.intersection(a)
        return s
        
l = read_union('./world_population.csv')
print(l)
```

### 19. Mediante operación de conjunto obtener los países que se encuentran en el conjunto del ejercicio 16 pero no en el conjunto del ejercicio 17.

```python
def read_union(name_file):
    with open(name_file) as csvfile:
        worlds = csv.reader(csvfile, delimiter=',')
        h = next(worlds)
        t = set()
        a = set()
        s = set()
        for i in worlds:
            if float(i[16]) > 0.2:
                t.add(i[2])
            if float(i[16]) < 0.4:
                a.add(i[2])
        s = a.symmetric_difference(a.intersection(t))
        return s
        
l = read_union('./world_population.csv')
print(l)
```
